# Graph Visualisation

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## 점선 그래프 

```python
data =  np.random.randn(50).cumsum()
data
```

```python
plt.plot(data)
plt.show()
```

##  여러그래프 그릴 준비 하기

```python
plt.subplot(1, 2, 1)
plt.subplot(1, 2, 2)
```

## 멀티 그래프 그리기 


```python
hist_data = np.random.randn(100)
scat_data = np.arange(30)
```

```python
plt.subplot(2, 2, 1)
plt.plot(data)

plt.subplot(2, 2, 2)
plt.hist(hist_data, bins=20)

plt.subplot(2, 2, 3)
plt.scatter(scat_data, np.arange(30) + 3 * np.random.rand(30))
```
